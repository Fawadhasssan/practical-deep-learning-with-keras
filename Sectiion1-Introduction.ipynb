# Learn to Predict the class

```python
from sklearn import svm
```

```python
clf = svm.SVC(gamma=0.01, C=100.)
```

```python
clf.fit(digits.data[:-1], digits.target[:-1])
```

```python
clf.predict(digits.data[-1:])
```
